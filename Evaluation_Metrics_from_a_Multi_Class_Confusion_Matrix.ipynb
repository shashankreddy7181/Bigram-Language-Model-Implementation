```python
import numpy as np
from sklearn.metrics import precision_score, recall_score

# confusion matrix
matrix = np.array([
    [5,10,5],
    [15,20,10],
    [0,15,10]
])

classes = ["Cat","Dog","Rabbit"]


y_true = []
y_pred = []

for i in range(len(classes)):
    for j in range(len(classes)):
        count = matrix[i, j]
        y_true.extend([i] * count)
        y_pred.extend([j] * count)

# Per-class precision and recall
class_precisions = precision_score(y_true, y_pred, average=None, labels=range(len(classes)))
class_recalls = recall_score(y_true, y_pred, average=None, labels=range(len(classes)))

for i, cls_name in enumerate(classes):
    print(f"{cls_name} -> Precision:{class_precisions[i]:.2f}, Recall:{class_recalls[i]:.2f}")

# Macro averages
macro_precision = precision_score(y_true, y_pred, average='macro')
macro_recall = recall_score(y_true, y_pred, average='macro')

# Micro averages
micro_precision = precision_score(y_true, y_pred, average='micro')
micro_recall = recall_score(y_true, y_pred, average='micro')

print(f"\nMacro Precision: {macro_precision:.2f}")
print(f"Macro Recall: {macro_recall:.2f}")
print(f"Micro Precision: {micro_precision:.2f}")
print(f"Micro Recall: {micro_recall:.2f}")
```
